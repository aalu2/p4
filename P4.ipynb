```python
#Answer of 1
def max_product(s):    
    inc = 0
    exc = 1    
    for i in s: 
          
        # Current max excluding i 
        new_exc = exc if exc>inc else inc
  
        # Current max including i 
        inc = exc * i 
        exc = new_exc
      
    # returning max of incl and excl 
    return (exc if exc>inc else inc)

 #Testing the code
print(max_product([10,3,1,9,2]))
print(max_product([]))
```

```python
#Answer of 2
def coords(fn, seq, lower, upper):
    return [[x, fn(x)] for x in seq if lower <= fn(x) <= upper]

# Testing the function here. ignore/remove the code below if not required
print(coords(lambda x: x ** 2, [-4, -2, 0, 1, 3], 1,9))
```

```python
#Answer of 3
lst = [-4, 1, -2, 0, 3]
print("Original List:", lst)
for j in range(len(lst)):
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            lst[i], lst[i + 1] = lst[i + 1], lst[i]
print("Sorted List:", lst)
```

```python
#Answer of 4
def f(arr,val):
    for i in arr:
        if(i==val):
            print("True")
            return
    
    print("False")
#Testing the code
arr = [1,2,3,4,5,6] 
f(arr,6)
```
